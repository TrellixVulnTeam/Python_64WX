# 4. 파이차트 그리기
## 4.1 파이차트
- 각 범주(Category)가 데이터에서 차지하는 비율을 나타내는데 사용
- `pie(x, labels)` 이용
    - x: 값 (값들을 100을 기준으로 비율을 계산해 크기 설정)
    - labels : 값들의 label
    - autopct: 조각내에 표시될 비율의 문자열 형식. '%fmt문자' 
        - fmt문자: f(실수), d(정수), %% (%)
            

```python
a = 10 
b =5.23


'%d - %.2f%%' % (a,b)
```

```python
f'{a},{b}'
```

```python
'{} - {}'.format(a,b)
```

```python
import matplotlib.pyplot as plt
```

```python
labels = ['pear','apple','melon','peach', 'berry']
amount = [20,10,40,5,60]
```

```python
plt.figure(figsize=(10,10))
plt.pie(amount,labels=labels, autopct="%.2f%%", 
        explode=[0,0.2,0,0,0.2], # explode:특정 값을 강조할때, 숫자만큼 빼준다.
        shadow=True,             # shadow: 차트 꾸미기, 입체적으로 보이게한다.
       textprops={'fontsize':15})# 폰트사이즈 조절.           

plt.show()
```

## .1 히스토그램
- 표로 된 도수 분포표를 그래프로 나타낸 것.
    - 도수분포표: 특정 그룹안에 값의 빈도를 나타낸 표
    - 주로 연속형 자료를 특정 구간으로 나눠 그 빈도를 확인한다.
        - 빈도 확인이나 **분포**를 볼때 사용한다.
    - X축: 계급(변수구간)의 대표값, Y축: 빈도수
- `hist(data [, bins=계급개수)` 메소드를 사용
    - data: 리스트형의 객체를 전달한다.

# 6. 상자그래프(Boxplot) 그리기
## 6.1 상자그래프란
- 값들의 중간값, 사분위수, 최대/최소값을 보여주어 값들의 분포를 확인 할때 사용한다.
- boxplot(x, whis=1.5)
    - x: 데이터셋
    - whis: 극단치 계산시 사용할 값(기본 1.5)
        - 극단치 : 1사분위와 3사분위에서 $IQR(3사분위수-1사분위수) * whis$ 범위 이상 떨어진 값들
