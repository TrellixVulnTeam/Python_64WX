# 4. 파이차트 그리기
## 4.1 파이차트
- 각 범주(Category)가 데이터에서 차지하는 비율을 나타내는데 사용
- `pie(x, labels)` 이용
    - x: 값 (값들을 100을 기준으로 비율을 계산해 크기 설정)
    - labels : 값들의 label
    - autopct: 조각내에 표시될 비율의 문자열 형식. '%fmt문자' 
        - fmt문자: f(실수), d(정수), %% (%)
            

```python
a = 10 
b =5.23


'%d - %.2f%%' % (a,b)
```

```python
f'{a},{b}'
```

```python
'{} - {}'.format(a,b)
```

```python
import matplotlib.pyplot as plt
```

```python
labels = ['pear','apple','melon','peach', 'berry']
amount = [20,10,40,5,60]
```

```python
plt.figure(figsize=(10,10))
plt.pie(amount,labels=labels, autopct="%.2f%%", 
        explode=[0,0.2,0,0,0.2], # explode:특정 값을 강조할때, 숫자만큼 빼준다.
        shadow=True,             # shadow: 차트 꾸미기, 입체적으로 보이게한다.
       textprops={'fontsize':15})# 폰트사이즈 조절.           

plt.show()
```

# 5 히스토그램 그리기
## .1 히스토그램
- 표로 된 도수 분포표를 그래프로 나타낸 것.
    - 도수분포표: 특정 그룹안에 값의 빈도를 나타낸 표
    - 주로 연속형 자료를 특정 구간으로 나눠 그 빈도를 확인한다.
        - 빈도 확인이나 **분포**를 볼때 사용한다.
    - X축: 계급(변수구간)의 대표값, Y축: 빈도수
- `hist(data [, bins=계급개수)` 메소드를 사용
    - data: 리스트형의 객체를 전달한다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(100, 10, 1000) #평균:100, 표준편차:10, 개수:1000 실수 난수
a= np.unique(data, return_counts=True) #고유값 + 각각의 개수
# a[0].min(), a[0].max()
a
```

```python
plt.bar(a[0], a[1])  #x:값, y:수량
```

```python
plt.hist(a[0])
plt.xticks([60,70,80,90,100,110,120,130,140,150])
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.hist(a[0], bins=100)

plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.hist(a[0], bins=50)
plt.xticks(range(66,140,2),rotation=90)
plt.show()
```

# 6. 상자그래프(Boxplot) 그리기
## 6.1 상자그래프란
- 값들의 중간값, 사분위수, 정상범위내에서 최대/최소값을 보여주어 값들의 분포를 확인 할때 사용한다.
- boxplot(x, whis=1.5)
    - x: 데이터셋
    - whis: 극단치 계산시 사용할 값(기본 1.5)
        - 극단치 : 1사분위와 3사분위에서 $IQR(3사분위수-1사분위수) * whis$ 범위 이상 떨어진 값들

```python
data = np.random.randint(1,100, 1000) # 1 ~ 99 1000개의 정수 난수
data[:5] = [200,250, -50, -100, -60]
data[:10]
```

```python
plt.figure(figsize=(7,7))
plt.boxplot(data)
plt.boxplot(data, whis=4) #whis 를 1.5(기본)초과 주면 정상범위를 늘리는 것.
#plt.boxplot(data, whis=0.3) #whis 를 1.5(기본)미만 주면 정상범위를 좁히는 것.
#plt.boxplot(data, vert=False) #vert 수직방향(기본-True), 수평방향(False)
plt.xticks([1],['x tick'])
plt.show()
```

## 상자그래프 활용
- tips_df 에서 tip의 분포를 상자그래프로 확인

```python
plt.figure(figsize=(7,7))
plt.boxplot(tips['tip'])
plt.xticks([1],['Tip'])
plt.show()
```

```python
tips['tip']
```

- 남녀별 tip의 분포글 하나의 boxplot에서 그려 분포 비교
- 한 그래프에서 여러 데이터셋의 boxplot을 그리기.
    - 2차원 배열형태의 구조로로 데이터를 제공
    - 행(axis=0) 별로 그래프를 그린다.

```python
male = tips[tips['sex']=='Male']['tip']
female = tips[tips['sex']=='Female']['tip']
```

```python
plt.figure(figsize=(7,7))
plt.boxplot([male, female])
plt.xticks([1,2], ['남','녀'])
plt.show()
```
